```python
import numpy as np
import matplotlib.pyplot as plt
import copy
from scipy.stats import multivariate_normal
from copy import deepcopy

np.random.seed(42)

%matplotlib inline
```

### Data Set

We will create some synthetic data where each data comes from one of the three gaussian distributions. Since, we initialize the means and covariances of these distribution, we know the underlying distribution of the data. When we implement k-Means and GMM, we can evaluate our performance against our knowledge of this underlying distribution. 

```python
def generate_data(num_data, means, covariances, weights):
    """ Creates a list of data points """
    
    num_clusters = len(weights)
    data = []
    for i in range(num_data):
        #  Use np.random.choice and weights to pick a cluster id
        k = np.random.choice(len(weights), 1, p=weights)[0]
        
        # Use np.random.multivariate_normal to create data from this cluster
        x = np.random.multivariate_normal(means[k], covariances[k])

        data.append(x)
    return data
```

Let's generate some data from the gaussian distributions with parameters specified below. Each gaussian is a multivariate gaussian distribution so for each gaussian you have a mean vector and a covariance matrix. My choice for these initial parameters is arbitrary. You can experiment with different parameters to obtain a different set. I have also given each gaussian distribution a weight (probability) - e.g. if one gaussian has probability 0.4, it means with probability 0.4 generate a data point that belongs to this distribution. This will be more helpful when we look at soft clustering using Gaussian Mixture Models where each data point has a probability distribution over class labels instead of a hard prediction. 

Using the above method, I will generate 100 data points and plot them.

```python
init_means = [
    [6, 3], # mean of cluster 1
    [3, 7], # mean of cluster 2
    [0, 2]  # mean of cluster 3
]
init_covariances = [
    [[.5, 0.], [0, .5]], # covariance of cluster 1
    [[.92, .38], [.38, .91]], # covariance of cluster 2
    [[.5, 0.], [0, .5]]  # covariance of cluster 3
]
init_weights = [1/4., 1/2., 1/4.]  # weights of each cluster

# Generate data
#np.random.seed(4)
data = generate_data(100, init_means, init_covariances, init_weights)
```

```python
plt.figure()
d = np.vstack(data)
plt.plot(d[:,0], d[:,1],'o')
plt.rcParams.update({'font.size':16})
plt.tight_layout()
```

### K-means Algorithm

k-means algorithm is an unsupervised learning algorithm. This is because we are given a training set $(x^{(1)}, ..., x^{(n)})$ and we don't have any labels $y^{(i)}$ associated with the data, so we want to group the data into clusters - a way to find out if there is any natural grouping in our training set. 

The k-means clustering algorithm is a __hard clustering__ method since we assign a label to each data point and have absolute faith that it belongs to the respective cluster. We don't define any probability distribution over the data points and hence a point that's close to 2 different centroids will go into the cluster whose centroid is relatively closer to the point, even though there's a chance it can go in either cluster. 

The algorithm is as follows:

1) Initialize cluster centroids $\mu_{1}, \mu_{2}, ... , \mu_{k}$ in $R^{d}$ randomly

2.1) For every i, set: $c^{(i)} = argmin_{j} ||x^{(i)} - \mu_{j}||^{2}$

2.2) For each j, set: $\mu_{j} = \frac{\sum_{i=1}^n 1(c^{(i)} = j)x^{i}}{\sum_{i=1}^n 1(c^{(i)} = j)}$

3) Repeat (2.1) and (2.2) until convergence


In the algorithm above, we define the hyperparameter k, which is the number of clusters. The $\mu_{j}$ is our current centroid position for $j = 1,...,k$. The initial guess for $\mu_{j}$ is random and then we follow step (2.1) and (2.2) to adjust the position of the centroids and reassign each data point a new cluster. 

Let's look at the code below and apply it on our data. 

```python
#Euclidean Distance function
def l2_dist(a, b,ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

```python
def k_means(data, centroids):
    
    min_data = np.min(data, axis=0)
    max_data = np.max(data, axis=0)
    
    x_min, x_max = min_data[0], max_data[0]
    y_min, y_max = min_data[1], max_data[1]
    
    _x = np.random.randint(x_min, x_max, size=centroids) # X coordinates of random centroids
    _y = np.random.randint(y_min, y_max, size=centroids) # Y coordinates of random centroids
    
    c = np.array(list(zip(_x, _y)), dtype=np.float32) # To store the value of centroids when it updates
    c_prev = np.zeros(c.shape)
    
    clusters = np.zeros(len(data)) # Cluster Lables(0, 1, 2)
    
    diff = l2_dist(c, c_prev) # distance between updated centroid and old centroid
    
    i_ = 0
    
    # stop when the centroids don't update anymore or no data point changes clusters
    max_iter = 20 # maximum plots to show
    while np.sum(diff) != 0:
        i_ += 1
        # Assigning each value to its closest cluster
        for i in range(len(data)):
            distances = l2_dist(data[i], c)
            cluster = np.argmin(distances)
            clusters[i] = cluster
            
        # Storing the old centroid values
        c_prev = deepcopy(c)
        
        # Finding the new centroids by taking the average value
        for i in range(centroids):
            points = [d[j] for j in range(len(data)) if clusters[j] == i]
            c[i] = np.mean(points, axis=0)
            
        diff = l2_dist(c, c_prev)
        
        if max_iter >= 0:
            plt.figure()
            plt.title("Centroids after iteration {}".format(i_))
            _data = np.vstack(data)
            plt.plot(_data[:,0], _data[:,1],'ko')
            plt.plot([c_prev[0][0],c_prev[1][0],c_prev[2][0]],[c_prev[0][1],c_prev[1][1],c_prev[2][1]], 'rs',markersize=12)
            #plt.plot([cc[0][0],cc[1][0],cc[2][0]],[cc[0][1],cc[1][1],cc[2][1]], 'yo',alpha=0.5,markersize=12)
            plt.rcParams.update({'font.size':16})
            plt.tight_layout()
            max_iter -= 1
    return c, clusters
```

```python
cent, points = k_means(d, 3)
```

Let's take a look at the final position of our clusters and the final grouping after the convergence of k-means algorithm.

The choice of random centroids can sometimes result in __poor convergence rate__, or convergence to __sub-optimal__ clustering. 

K-means algorithm can get stuck easily in local minima. There are several things we can do to avoid getting stuck in the local minima:
- Select good random centroids using a heuristic (e.g. least similar to any existing means)
- Try out multiple starting points (important)
- Initialize with the results of another method

Despite anything, k-means algorithm is guaranteed to converge. We have to make sure that the convergence is not to a sub-optimal solution. 

### Gaussian Mixture Model and Expectation Maximization Algorithm

__Soft Clustering__: instead of assigning each data point a particular cluster, we assign each data point a probability distribution over clusters. $ P(clusterIdx | x) $ instead of $clusterIdx = f(x)$.

There are 3 reasons we may want to perform clustering probabilistically:
- Handle missing data
- Tune Hyperparameters
- Generate new data points

GMM is a probabilistic model that consists of several gaussians. Each gaussian explains one cluster and the density of each data poit is the weighted sum of different gaussian densities. For example, if we have 3 gaussians, then: $P(x|\theta) = \pi_{1}N(x|\mu_{1},\sum_{1}) + \pi_{2}N(x|\mu_{2},\sum_{2}) + \pi_{3}N(x|\mu_{3},\sum_{3})$

In this case, the model parameters are: $\theta = (\pi_{1}, \pi_{2}, \pi_{3}, \mu_{1}, \mu_{2}, \mu_{3}, \sum_{1}, \sum_{2}, \sum_{3})$

If we optimize for the parameters using Maximum Liklehood Estimation, we'll find that it's not possible to find the MLE of the parameters in a closed form. This is why using Stochastic Gradient Descent (SGD) for MLE is not often useful. 

So what can we do? We introduce a __latent variable__. Latent simply means hidden so in our case it's the unobserved variable that we may want in order to make optimizing GMM easier. 

For each data point, we don't know which gaussian it comes from so that could serve as a latent variable. Let's call this latent variable __source__ $= z^{(i)}$. 

Now, the GMM problem is as follows: Given a training set $(x^{(1)},...,x^{(n)})$, we wish to model the data by specifiying a joint distribution $p(x^{(i)},z^{(i)}) = p(x^{(i)}|z^{(i)})p(z^{(i)})$. Here $z^{(i)}$ ~ $Multinomial(\phi)$ (where $\phi_{j} >= 0, \sum_{j=1}^{k}\phi_{j} = 1)$, the parameters $\phi_{j}$ gives $p(z^{(i)} = j)$, and $x^{(i)}|z^{(i)} = j$ ~ $Normal(\mu_{j}, \sum_{j})$. We let k denote the number of values that the $z^{(i)}$'s can take on. 

Thus, our model means that each $x^{(i)}$ was generated by randomly choosing $z^{(i)}$ from $(1,...,k)$, and then $x^{(i)}$ was drawn from one of k Gaussians depending on $z^{(i)}$. 


The random variable $z^{(i)}$ indicate  which of the k Gaussians each $x^{(i)}$ had come from. Note that if we knew what the $z^{(i)}$'s were, the MLE problem would have been easy. So two essential questions:

- How can we estimate the model parameters $\theta$? The answer is if we know the sources $z^{(i)}$'s then we can solve the MLE easily and get the estimate of our model parameters. However, in general we don't know the sources. This leads to the second questions.

- How can we estimate the sources $z^{(i)}$'s? The answer is if we know the parameters of our model, then we can easily estimate the sources e.g. given $p(x|z=1,\theta) = Normal(-2,1)$, find $p(z=1|x,\theta)$. So we will get, $p(z=1|x,\theta) = \frac{p(x|z=1,\theta)p(z=1|\theta)}{Z}$

It is a chicken and egg problem! 
- Need Gaussian parameters to estimate the sources
- Need sources to estimate the Gaussian parameters

So, what do we do? This is where the Expectation Maximization (EM) algorithm comes in. The EM algorithm is an iterative algorithm that has two main steps. Informally, in the E step, it tries to guess the values of the sources $z^{(i)}$'s and in the M step, it updates the model parameters of our model based on our guesses. Since in the M step we are pretending that the guesses in the first part were correct, the maximization becomes easy. 

EM Algorithm:

1) Start with k randomly placed Gaussian parameters $\theta$

2) E-step: for each point compute the posterior probability $p(z = c | x^{(i)},\theta)$

3) M-step: update the Gaussian parameters $\theta$ to fit points assigned to them. 

4) Repeat step 2 and 3 until convergence

Let's optimize GMM on our data above using EM algorithm. 

```python
def log_sum_exp(Z):
    """ Compute log(\sum_i exp(Z_i)) for some array Z."""
    return np.max(Z) + np.log(np.sum(np.exp(Z - np.max(Z))))

def loglikelihood(data, weights, means, covs):
    """ Compute the loglikelihood of the data for a Gaussian mixture model with the given parameters. """
    num_clusters = len(means)
    num_dim = len(data[0])
    
    ll = 0
    for d in data:
        Z = np.zeros(num_clusters)
        for k in range(num_clusters):
            
            # Compute (x-mu)^T * Sigma^{-1} * (x-mu)
            delta = np.array(d) - means[k]
            exponent_term = np.dot(delta.T, np.dot(np.linalg.inv(covs[k]), delta))
            
            # Compute loglikelihood contribution for this data point and this cluster
            Z[k] += np.log(weights[k])
            Z[k] -= 1/2. * (num_dim * np.log(2*np.pi) + np.log(np.linalg.det(covs[k])) + exponent_term)
            
        # Increment loglikelihood contribution of this data point across all clusters
        ll += log_sum_exp(Z)
        
    return ll
```

```python
def EM(data, init_means, init_covariances, init_weights, maxiter=1000, thresh=1e-4):
    
    means = init_means[:]
    covariances = init_covariances[:]
    weights = init_weights[:]
    
    num_data = len(data)
    num_dim = len(data[0])
    num_clusters = len(means)
    
    resp = np.zeros((num_data, num_clusters))
    ll = loglikelihood(data, weights, means, covariances)
    ll_trace = [ll]
    
    for i in range(maxiter):
        if i % 5 == 0:
            print("Iteration %s" % i)
            
        # E-step: compute responsibilities
        # Update resp matrix so that resp[j, k] is the responsibility of cluster k for data point j.
        for j in range(num_data):
            for k in range(num_clusters):
                resp[j, k] =  weights[k] * multivariate_normal.pdf(data[j], mean=means[k], cov=covariances[k])
        row_sums = resp.sum(axis=1)[:, np.newaxis]
        resp = resp / row_sums # normalize over all possible cluster assignments

        # M-step
        counts = np.sum(resp, axis=0)
        
        for k in range(num_clusters):
            
            # Update the weight for cluster k using the M-step update rule for the cluster weight
            weights[k] = counts[k]
            
            # Update means for cluster k using the M-step update rule for the mean variables.
            weighted_sum = 0
            for j in range(num_data):
                weighted_sum += data[j] * resp[j,k]
            means[k] = weighted_sum / weights[k]
            
            # Update covariances for cluster k using the M-step update rule for covariance variables.
            weighted_sum = np.zeros((num_dim, num_dim))
            for j in range(num_data):
                weighted_sum += np.outer(data[j] - means[k],data[j] - means[k]) * resp[j,k]

            covariances[k] = weighted_sum / weights[k]

        
        
        # Compute the loglikelihood at this iteration
        ll_latest = loglikelihood(data, weights, means, covariances)
        ll_trace.append(ll_latest)
        
        # Check for convergence in log-likelihood and store
        if (ll_latest - ll) < thresh and ll_latest > -np.inf:
            break
        ll = ll_latest
    
    if i % 5 != 0:
        print("Iteration %s" % i)
    weights = weights / sum(weights)
    out = {'weights': weights, 'means': means, 'covs': covariances, 'loglik': ll_trace, 'resp': resp}

    return out
```

```python
# Initialization of parameters
chosen = np.random.choice(len(data), 3, replace=False)
initial_means = [data[x] for x in chosen]
initial_covs = [np.cov(data, rowvar=0)] * 3
initial_weights = [1/3.] * 3

# Run EM 
results = EM(data, initial_means, initial_covs, initial_weights)
```

```python
import matplotlib.mlab as mlab
def plot_contours(data, means, covs, title):
    plt.figure()
    plt.plot([x[0] for x in data], [y[1] for y in data],'ko')

    delta = 0.025
    k = len(means)
    x = np.arange(-2.0, 7.0, delta)
    y = np.arange(-2.0, 7.0, delta)
    X, Y = np.meshgrid(x, y)
    col = ['green', 'red', 'blue']
    for i in range(k):
        mean = means[i]
        cov = covs[i]
        sigmax = np.sqrt(cov[0][0])
        sigmay = np.sqrt(cov[1][1])
        sigmaxy = cov[0][1]/(sigmax*sigmay)
        Z = mlab.bivariate_normal(X, Y, sigmax, sigmay, mean[0], mean[1], sigmaxy)
        plt.contour(X, Y, Z, colors = col[i])
        plt.title(title)
    plt.rcParams.update({'font.size':16})
    plt.tight_layout()
```

```python
# Parameters after initialization
plot_contours(d, initial_means, initial_covs, 'Initial clusters')
```

```python
# Parameters after running EM to convergence
plot_contours(d, results['means'], results['covs'], 'Final clusters')
```
